Load up the wheat seeds dataset into a dataframe. We've stored a copy in the Datasets directory.

```python
# load wheat data
df = pd.read_csv('Datasets\wheat.data')
df.head()
```

Create a 2d scatter plot that graphs the `area` and `perimeter` features:

```python
def plot_scatter(frame, pyp, x, y,r,marker, title, xlab, ylab):
    frame.plot.scatter(x=x,y=y,c='r',marker='*')
    pyp.suptitle(title)
    pyp.xlabel(xlab)
    pyp.ylabel(ylab)
    pyp.show()
    
plot_scatter(df, plt, 'area', 'perimeter', 'r', '*', 'area vs perimeter', 'area', 'perimeter')


#df.plot.scatter(x='area',y='perimeter',c='r',marker='*')
#plt.suptitle('area Vs perimeter')
#plt.xlabel('area')
#plt.ylabel('perimeter')
#plt.show()
```

Create a 2d scatter plot that graphs the `groove` and `asymmetry` features:

```python
plot_scatter(df, plt, 'groove', 'asymmetry', 'groove vs asymmetry', 'groove', 'asymmetry')
```

Create a 2d scatter plot that graphs the `compactness` and `width` features:

```python
plot_scatter(df, plt, 'compactness', 'width', 'compactness vs width', 'compactness', 'width')
```
