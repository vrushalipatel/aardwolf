```python
def plot_scatter(frame, pyp, x, y, title, xlab, ylab):
    frame.plot.scatter(x=x,y=y)
    pyp.suptitle(title)
    pyp.xlabel(xlab)
    pyp.ylabel(ylab)
    pyp.show()
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
df=pd.read_excel("Datasets/Concrete_Data.xls", 'Sheet1')
matplotlib.style.use('ggplot')
df.columns = ['cement', 'slag', 'fly-ash', 'water', 'plast', 'coarse-agg', 'fine-agg', 'age', 'strength']

plot_scatter(df, plt, 'cement', 'strength', 'cement vs strength', 'cement', 'strength')
plot_scatter(df, plt, 'slag', 'strength', 'slag vs strength', 'slag', 'strength')
plot_scatter(df, plt, 'fly-ash', 'strength', 'fly-ash vs strength', 'fly-ash', 'strength')
plot_scatter(df, plt, 'water', 'strength', 'water vs strength', 'water', 'strength')
plot_scatter(df, plt, 'plast', 'strength', 'plast vs strength', 'plast', 'strength')
plot_scatter(df, plt, 'coarse-agg', 'strength', 'coarse-agg vs strength', 'coarse-agg', 'strength')
plot_scatter(df, plt, 'fine-agg', 'strength', 'fine-agg vs strength', 'fine-agg', 'strength')
plot_scatter(df, plt, 'age', 'strength', 'age vs strength', 'age', 'strength')

#
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd
df=pd.read_excel("Datasets/Concrete_Data.xls", 'Sheet1')
matplotlib.style.use('ggplot')
df.columns = ['cement', 'slag', 'fly-ash', 'water', 'plast', 'coarse-agg', 'fine-agg', 'age', 'strength']

fig=plt.figure()
plt.suptitle('Cement X Water X Concrete Strenght')
ax = fig.add_subplot(111, projection = '3d')
ax.set_xlabel('Cement Content')
ax.set_ylabel('Water Content')
ax.set_zlabel('Concrete Content')
ax.scatter(df.cement,df.water,df.strength, c='g',marker ='o')
plt.show()
```
